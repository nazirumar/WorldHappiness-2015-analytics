```python
import opendatasets as pds
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams["figure.figsize"] = (20, 10)
```

```python
pds.download('https://www.kaggle.com/unsdsn/world-happiness')
```

```python
df =pd.read_csv('2015.csv')
df.head(3)
```

```python
df.info()
```

```python
df.describe()
```

```python
ndf = df.drop(['Country', 'Region'], axis=1)
```

```python
df=df.rename(columns={'Happiness Rank':'Happiness', 'Happiness Score':'Happiness_Score'})
```

```python

sns.heatmap(ndf, linecolor='r')
```

```python
plt.style.use('ggplot')

ax = sns.barplot(data=df, y='Happiness_Score', x='Region')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
plt.legend(df['Region'])
```

```python
ax =sns.histplot(data=df, x='Country', y='Freedom' )
ax.set_xticklabels (ax.get_xticklabels(), rotation=30)
plt.legend(df['Country'])
```

```python

ax =sns.pairplot(data=df,  height=3.5)
# ax.set_xticklabels (ax.get_xticklabels(), rotation=30)
plt.legend(df['Country'])
```
